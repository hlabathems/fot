```python
#!/usr/bin/env python
'''
2014 April 5
ARW & LAM, JPL/Caltech
Full Of Time

Updated
2015 April 6
CMG JPL Intern
lcolens
'''
import numpy as np
import random
import triangle
import matplotlib
from astropy.io import fits
import os.path
import csv

####################################################################
# From hubble_observation_schedule.py
def hubble_obs_sched(num_orbits):
  t_sample=[0.] #initialize time sampling array
  t=[0.] #initialize observation time array
  T_orbit=90. #in minutes
  frac_obs=0.6   #fraction of orbit where source is visible
  t_int=5. # integration time in minutes
  time_uncertainty = .5 #uncertainty on when the satellite is ready to observe in minutes
  min_per_day=1./(60.*24) #convert from minutes to days
  bail=0 #flag for bailing out of the time sampling loop
  while (bail==0):
    t_val=t_sample[len(t_sample)-1]+t_int+(random.gauss(0.,time_uncertainty)) #in minutes
    t_sample.append(t_val)
    if(np.mod(t_val,T_orbit)<frac_obs*T_orbit):
      t.append(t_val)
    if(t_val>num_orbits*T_orbit): bail=1
  t=np.array(t)*min_per_day
  return t
####################################################################

####################################################################
# From fot_hubble_sim.py, parser converted to object
class Parameters:
    def __init__(self):
        self.delay = 1.0    # Delay between images,type=float 
        self.delay_prior = 1.0    # Delay prior for emcee,type=float 
        self.delay_prior_min = 0.0    # Delay prior minimum acceptable value for emcee,type=float
        self.delay_prior_max = 2.0    # Delay prior maximum acceptable value for emcee,type=float 
        self.delta_mag = .5    # Difference in magntiude between images,type=float
        self.delta_mag_prior = 1.    # Difference in magnitude between images prior for emcee,type=float
        self.delta_mag_prior_min = 1.    # Difference in magnitude between images prior minimum acceptable value for emcee,type=float
        self.delta_mag_prior_max = 1.    # Difference in magnitude between images prior maximum acceptable value for emcee,type=float
        self.sigma = 4.    # Quasar light curve sigma parameter,type=float
        self.sigma_prior = 4.    # Quasar light curve sigma prior for emcee,type=float
        self.sigma_prior_min = 3.8    # Quasar light curve sigma prior minimum acceptable value for emcee,type=float
        self.sigma_prior_max = 4.2    # Quasar light curve sigma prior maximum acceptable value for emcee,type=float
        self.tau = 100.    # Quasar light curve tau parameter,type=float
        self.tau_prior = 1.    # Quasar light curve prior for emcee,type=float
        self.tau_prior_min = 1.    # Quasar light curve prior minimum acceptable value for emcee,type=float
        self.tau_prior_max = 1.    # Quasar light curve prior maximum acceptable value for emcee,type=float
        self.avg_mag = 19.    # Quasar light curve average magnitude,type=float
        self.avg_mag_prior = 1.    # Quasar light curve magnitude prior for emcee,type=float
        self.avg_mag_prior_min = 1.    # Quasar light curve magnitude prior minimum acceptable value for emcee,type=float
        self.avg_mag_prior_max = 1.    # Quasar light curve magnitude prior maximum acceptable value for emcee,type=float
        self.redshift = 1.    # Quasar redshift,type=float
        self.numorbits = 47    # Number of Hubble orbits to be used,type=int
        self.photometric_error = 1.    # Photometric error uncertainty (e.g. 0.02) [mag],type=float
        self.output_tag = "String"    # Output tag, in quotes,type=str
####################################################################
```

```python
# Gather time stamps from observational data, mjd less than 57008 does not include test data for our campaign
mjd_list = []
def step(ext, dirname, names):
    ext = ext.lower()
 
    for name in names:
        if name.lower().endswith('fits'):
            fits_file_name = (os.path.join(dirname, name))
            hdulist=fits.open(fits_file_name)
            if hdulist[0].header['MJD-OBS'] > 57008:
                mjd_list.append(hdulist[0].header['MJD-OBS'])

os.path.walk('../../DataCorrected/he045-1223_wcs_corrected/', step, 'fits')
mjd_sorted = sorted(mjd_list)
```
